```python
# Importando as biliotecas necessárias
import pandas as pd
import numpy as np
```

```python
# Obtendo o dataset
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Machine-Learning-Algorithms/master/Dataset/Iris.csv'

dados = pd.read_csv( url )
```

```python
# Quantidade de linhas e colunas
dados.shape
```

```python
# Visualizando alguns registros
dados.head()
```

```python
# Mudança da semente aleatória randômica para gerar os resultados em várias execuções
np.random.seed( 2345 )
```

```python
# 150 ( são 150 linhas no dataset ) amostras entre 0 e 1 com reposição, probabilidades equivalentes
amostra = np.random.choice( a=[0,1], size=150, replace=True, p=[0.7, 0.3] )
# Hiperparâmetros -> ( a= espaço amostral ), ( size= tamanho, quantidade de elementos da amostra ), 
#                    ( replace= com reposição, quando gerar o 0 ou 1, informa que pode gerar novamente, senão, gera 0 por exemplo e depois não pode mais gerar...teremos problema ), 
#                    ( p= probabilidades de geração de cada valor, não quer dizer percentual dos valores mas a probabilidade daquela valor ser gerado )
```

```python
# Verificar o tamanho da amostra
len( amostra )
```

```python
# Criar o campo amostra no dataset com as amostras criadas randômicamente
dados['amostra'] = amostra
dados
```

```python
# Verificar tamanho da amostra por valores
print( len( amostra[ amostra == 1 ] ) )
print( len( amostra[ amostra == 0 ] ) )
```
