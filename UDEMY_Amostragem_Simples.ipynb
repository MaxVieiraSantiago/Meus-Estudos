```python
# Importando as biliotecas necess√°rias
import pandas as pd
import numpy as np
```

```python
# Obtendo o dataset
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Machine-Learning-Algorithms/master/Dataset/Iris.csv'

dados = pd.read_csv( url )
```

```python
# Quantidade de linhas e colunas
dados.shape
```

```python
# Visualizando alguns registros
dados.head()
```

```python
# Mudan√ßa da semente aleat√≥ria rand√¥mica para gerar os resultados em v√°rias execu√ß√µes
np.random.seed( 2345 )
```

```python
# 150 ( s√£o 150 linhas no dataset ) amostras entre 0 e 1 com reposi√ß√£o, probabilidades equivalentes
amostra = np.random.choice( a=[0,1], size=150, replace=True, p=[0.7, 0.3] )
# Hiperpar√¢metros -> ( a= espa√ßo amostral ), ( size= tamanho, quantidade de elementos da amostra ), 
#                    ( replace= com reposi√ß√£o, quando gerar o 0 ou 1, informa que pode gerar novamente, sen√£o, gera 0 por exemplo e depois n√£o pode mais gerar...teremos problema ), 
#                    ( p= probabilidades de gera√ß√£o de cada valor, n√£o quer dizer percentual dos valores mas a probabilidade daquela valor ser gerado )
```

```python
# Verificar o tamanho da amostra
len( amostra )
```

```python
# Criar o campo amostra no dataset com as amostras criadas rand√¥micamente
dados['amostra'] = amostra
dados
```

```python
# Verificar tamanho da amostra por valores
print( len( amostra[ amostra == 1 ] ) )
print( len( amostra[ amostra == 0 ] ) )
```
