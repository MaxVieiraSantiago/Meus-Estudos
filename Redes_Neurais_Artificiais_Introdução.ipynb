```python
# N -> amostras
# D_in -> dimensão da camada de entrada
# H -> dimensão da camada escondida
# D_out -> dimensão da camada de saída

# x -> observações de entrada
# y -> alvo
# w1, w2 -> matrizes de peso

N, D_in, H, D_out = 64, 1000, 100, 10

# Gerando um numpy array de N posições para x, y
x, y = randn( N, D_in ), randn( N, D_out )

w1, w2 = randn( D_in, H ), randn( H, D_out )
```

```python
for t in range(2000):
  # calcular o valor da camada escondida ( neste caso a função sogmóide )
  h = 1 / ( 1 + np.exp( -x.dot( w1 ) )  ) 
  # numpy.dot() realiza a multiplicação de matrizes ( dimensões das duas tem que ser iguais Linha e Coluna )
  # no caso acima multiplicando a matriz x pela matriz de pesos w1
   
  # o u predito é resultado da multiplicação da matriz de valores da camada escondida (h) pela matriz de pesos w2
  y_pred = h.dot( w2 )

  # A função de perda é o somatório dos quadrados das diferenças entre a matriz de valores preditos (y_pred) pela matriz de observações (y)
  loss = np.square( y_pred - y ).sum()
  # print( t, loss )

  grad_y_pred = 2 * ( y_pred - y )
  grad_w2     = h.T.dot( grad_y_pred )
  grad_h      = grad_y_pred.dot( w2.T )
  grad_w1     = x.T.dot( grad_h * h * ( 1 - h ) ) 

  w1 -= 1e-4 * grad_w1
  w2 -= 1e-4 * grad_w2
```
