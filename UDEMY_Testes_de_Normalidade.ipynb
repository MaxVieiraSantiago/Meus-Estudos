```python
# Importação das bibliotecas necessárias
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# Criar uma variável com os dados de uma distribuição normal utilizano o rvs ( 1000 elementos )
dados = norm.rvs( size=1000 )
dados
```

```python
# Histograma
plt.hist( dados, bins= 30 )
plt.title( 'Dados' );
```

```python
# Geração do gráfico para verificar se a distruibição é normal
fig, ax = plt.subplots()
stats.probplot( dados, fit=True, plot=ax )
plt.show()
```

```python
# Teste de Shapiro
# Segundo argumento é o valor de p, não há com rejeitar a hipótese nula
stats.shapiro( dados )
```

## **Dados Não Normalizados**

```python
dados_02 = skewnorm.rvs( 4, size=1000 )
```

```python
# Histograma
plt.hist( dados_02, bins=30 )
```

```python
# Geração de gráfico para verificar se os dados são normais
fig, ax = plt.subplots()
stats.probplot( dados_02, fit=True, plot=ax )
plt.show()
```

```python
# Teste de Shapiro
# Segundo argumento é o valor de p, não há com rejeitar a hipótese nula
stats.shapiro( dados_02 )
```
