```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
print( 'cancer.keys(): \n{}'.format( cancer.keys() ) )
```

```python
print( 'shape of cancer data: {}'.format( cancer.data.shape ) )
```

```python
import numpy as np
print("Sample counts per class:\n{}".format( { n: v for n, v in zip( cancer.target_names, np.bincount( cancer.target ) ) } ) )
```

```python
import pandas as np
dados = np.DataFrame( cancer.data, columns=cancer.feature_names )
dados['target'] = cancer.target
dados
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = dados.drop( 'target', axis= 1 )
y = dados['target']
X_train, X_teste, y_train, y_teste = train_test_split( X, y, stratify=dados['target'], random_state=66 )
```

```python
acuracia_treinamento = []
acuracia_teste = []

# Tentar de 1 a 10 vizinhos
vizinhos = range( 1, 11 )

for vizinho in vizinhos:
  clf = KNeighborsClassifier( n_neighbors=vizinho )
  clf.fit( X_train, y_train )

  # Registrar a acurácia no Treinamento
  acuracia_treinamento.append( clf.score( X_train, y_train ) * 100 )

  # Registar a acurácia no Teste
  acuracia_teste.append( clf.score( X_teste, y_teste ) * 100 )
```

```python
import matplotlib.pyplot as plt

plt.plot( vizinhos, acuracia_treinamento, label='Acurácia no Treinamento' )
plt.plot( vizinhos, acuracia_teste, label='Acurácia no Teste' )
plt.ylabel( 'Acurácia' )
plt.xlabel( 'Vizinhos' )
plt.legend()
```
