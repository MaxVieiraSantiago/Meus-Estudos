# **NumPy**
**NumPy** é um dos pacotes fundamentais para computação científica em **Python**. Ele contém funcionalidades para matrizes multidimensionais, funções matemáticas de alto nível como operações de álgebra linear e a transformada de Fourier, e pseudo geradores de números.
No scikit-learn, o array **NumPy** é a estrutura de dados fundamental. **Scikit-learn** recebe dados na forma de matrizes **NumPy**. Todos os dados que você estiver usando terão que ser convertidos para uma matriz **NumPy**. A principal funcionalidade do **NumPy** é a classe **ndarray**, uma matriz multidimensional (n-dimensional). Todos os elementos da matriz devem ser do mesmo tipo. Uma matriz **NumPy** se parece com isso:

```python
import numpy as np

x = np.array( [ [ 1, 2, 3], [ 4, 5, 6 ] ] )

print( 'x:\n{}'.format(x) )
```

# **Modelando - Dados de Treinamento e Dados de Teste**

```python
from sklearn.model_selection import train_test_split

X = dados_iris['data']
y = dados_iris['target']

X_treino, X_teste, y_treino, y_teste = train_test_split( X, y, random_state = 0 )

# 75% dos dados
print( 'X_treino: {}'.format( X_treino.shape ) ) 
print( 'y_treino: {}'.format( y_treino.shape ) )

print( '\n' )

# 25% dos dados
print( 'X_teste:  {}'.format( X_teste.shape ) ) 
print( 'y_teste:  {}'.format( y_teste.shape ) )
```

# **Primeiro - Observar os Dados**

```python
# Criar um dataframe dos dados de treinamento ( X_treino )
dados_df = pd.DataFrame( X_treino, columns = dados_iris['feature_names'] )

# Criar um scatter_matrix do dataframe, coloriro por y_treino
grafico = pd.plotting.scatter_matrix( dados_df, 
                                      c=y_treino, 
                                      figsize=(20,20), 
                                      marker='o', 
                                      hist_kwds={'bins':20}, 
                                      s=60, 
                                      alpha=.8 )
```

# **Primeiro Modelo - K_Nearest Neighbors ( KNN )**

Agora podemos começar a construir o modelo de aprendizado de máquina real. Existem muitas classificações algoritmos no **scikit-learn** que poderíamos usar. Aqui vamos usar classificador **k-Nearest Neighbors ( KNN )** , que é fácil de entender. A construção deste modelo consiste apenas em armazenar o conjunto de treinamento. Para fazer uma previsão para um novo ponto de dados, o algoritmo encontra o ponto no conjunto de treinamento que está mais próximo do novo ponto. Em seguida, atribui o rótulo deste ponto de treinamento para o novo ponto de dados.

O **k** em **k** vizinhos mais próximos significa que, em vez de usar apenas o vizinho mais próximo para o novo ponto de dados, podemos considerar qualquer número fixo **k** de vizinhos no treinamento (por exemplo, os três ou cinco vizinhos mais próximos). Então, podemos fazer uma previsão usando a classe majoritária entre esses vizinhos. Todos os modelos de aprendizado de máquina no **scikit-learn** são implementados em suas próprias classes, que são chamadas de classes *Estimator*. O algoritmo de classificação dos **k-NN** é implementado na classe **NeighborsClassifier** no módulo **neighbors**. Antes
de usar o modelo, precisamos instanciar a classe em um objeto. Isto é quando formos definir quaisquer parâmetros do modelo. O parâmetro mais importante do **KNeighborsClassifier** é o **número de vizinhos** **( n_neighbors )**, que definiremos como 1:

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier( n_neighbors=1 )
knn.fit( X_treino, y_treino )
```

#**Realizando Predições**

```python
X_features = np.array( [ [ 5, 2.9, 1, 0.2 ] ] )
print( 'X_features: {}'.format( X_features.shape ) )
```

```python
predicao = knn.predict( X_features )
print( 'Predição: {}'.format(predicao) )
print( "Predição 'target_name': {}".format( dados_iris['target_names'][predicao] ) )
```

# **Avaliando o Modelo**

```python
y_pred = knn.predict( X_teste )

print( 'Predições no conjunto de teste: {}'.format( y_pred ) )
```

```python
# Pontuação no conjunto de testes
print( 'Acertos nos testes: {:.2f}%'.format( np.mean( y_pred == y_teste ) * 100 ) ) 
```

```python
# Pontuação do método do objeto knn
print( 'Acertos nos testes: {:.2f}%'.format( knn.score( X_teste, y_teste ) * 100 ) )
```
